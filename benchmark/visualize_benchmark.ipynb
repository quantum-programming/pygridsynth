## Moderate Precision

```python
import matplotlib.pyplot as plt
import json
import numpy as np

fig = plt.figure(figsize = (8, 4))

# Loading runtime stats based on laptop
t_pygridsynth_data = json.load(open("./data/pygridsynth_runtime.data", "r"))["runtime_data"]
t_newsynth_data = json.load(open("./data/newsynth_runtime.data", "r"))["runtime_data"]
eps_list = json.load(open("./data/newsynth_runtime.data", "r"))["eps_list"]

plt.plot(eps_list, np.median(t_pygridsynth_data, axis = 0), "-o", label = "pygridsynth")
plt.plot(eps_list, np.median(t_newsynth_data, axis = 0), "-o", label = "newsynth")

plt.xscale("log")
plt.xticks(10**np.linspace(-20, 0, 5), 10**np.linspace(-20, 0, 5))
plt.xlim(xmin=1e-21, xmax = 1e-0)
plt.ylim(ymax = 0.055, ymin = -0.003)

plt.legend()
plt.xlabel("epsilon")
plt.ylabel("runtime [sec]")

plt.grid()
plt.title(f"Clifford+T synthesis, average over random 20 angles")
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np

fig = plt.figure(figsize = (8, 4))

t_pygridsynth_data = json.load(open("./data/pygridsynth_runtime_pi_128.data", "r"))["runtime_data"]
t_newsynth_data = json.load(open("./data/newsynth_runtime_pi_128.data", "r"))["runtime_data"]
eps_list = json.load(open("./data/pygridsynth_runtime_pi_128.data", "r"))["eps_list"]

plt.plot(eps_list, np.median(t_pygridsynth_data, axis = 0), "-o", label = "pygridsynth")
plt.plot(eps_list, np.median(t_newsynth_data, axis = 0), "--o", label = "newsynth")
plt.plot(eps_list_newsynth, t_newsynth_list,"--s", label = "newsynth, paper")

plt.xscale("log")
plt.xticks(10**np.linspace(-100, -10, 10), 10**np.linspace(-100, -10, 10))
plt.xlim(xmin=1e-101, xmax = 1e-1)
plt.ylim(ymax = 0.5)

plt.legend()
plt.xlabel("epsilon")
plt.ylabel("runtime [sec]")

plt.grid()
plt.title(f"Clifford+T synthesis of pi/128")
```
